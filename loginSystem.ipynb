```python
from tkinter import *
import os

def register_user():
    user_name_info = user_name.get()
    passward_info = passward.get()
    
    file = open(user_name_info,'w')
    file.write(user_name_info+'\n')
    file.write(passward_info)
    file.close()
    
    user_name_entry.delete(0,END)
    passward_entry.delete(0,END)
    
    Label(window1,text='Registration Successfull',fg='green',font=('calibri',11)).pack()

def register():
    global window1
    global user_name
    global passward
    global user_name_entry
    global passward_entry
    
    window1 = Toplevel(window)
    window1.geometry('300x250')
    window1.title('Register')
    
    user_name=StringVar()
    passward=StringVar()
    
    Label(window1,text='Please enter details').pack()
    Label(window1,text='').pack()
    Label(window1,text='Username').pack()
    user_name_entry=Entry(window1,textvariable=user_name)
    user_name_entry.pack()
    Label(window1,text='').pack()
    Label(window1,text='Passward').pack()
    passward_entry=Entry(window1,textvariable=passward)
    passward_entry.pack()
    Label(window1,text='').pack()
    Button(window1,text='Register',height=2,width=20,command=register_user).pack()
def delete1():
    window3.destroy()
    
def login_success():
    global window3
    window3 = Toplevel(window)
    window3.geometry('300x250')
    window3.title('Success')
    Label(window3,text='Login Success!').pack()
    Button(window3,text='OK',command=delete1).pack()

def delete2():
    window4.destroy()
    
def passward_not_recognized():
    global window4
    window4 = Toplevel(window)
    window4.geometry('300x250')
    window4.title('Passward Incorrect')
    Label(window4,text='Passward not matched!').pack()
    Button(window4,text='OK',command=delete2).pack()
    
def delete3():
    window5.destroy()
    
def user_not_found():
    global window5
    window5 = Toplevel(window)
    window5.geometry('300x250')
    window5.title('User Incorrect')
    Label(window5,text='User Not Found!').pack()
    Button(window5,text='OK',command=delete3).pack()
    
def login_verify():
    user_name1 = user_name_verify.get()
    passward1 = passward_verify.get()
    
    user_name_entry1.delete(0,END)
    passward_entry1.delete(0,END)
    
    list_of_files = os.listdir()
    if user_name1 in list_of_files:
        file1 = open(user_name1,'r')
        verify = file1.read().splitlines()
        if passward1 in verify:
            login_success()
        else:
            passward_not_recognized()
    else:
        user_not_found()
    
def login():
    global window2
    global user_name_verify
    global passward_verify
    global user_name_entry1
    global passward_entry1
    window2 = Toplevel(window)
    window2.geometry('300x250')
    window2.title('Login')
    
    Label(window2,text='Please enter details').pack()
    Label(window2,text='').pack()
    
    user_name_verify = StringVar()
    Label(window2,text='Username').pack()
    user_name_entry1=Entry(window2,textvariable=user_name_verify)
    user_name_entry1.pack()
    
    passward_verify = StringVar()
    Label(window2,text='Passward').pack()
    passward_entry1=Entry(window2,textvariable=passward_verify)
    passward_entry1.pack()
    Label(window2,text='').pack()
    Button(window2,text='Login',height=2,width=20,command=login_verify).pack()

def main_func():
    global window
    
    window = Tk()
    window.geometry('300x200')
    window.title('LoginSystem')
    Label(text='Notes 1.0',bg='grey').pack()
    Label(text='').pack()
    Button(text='Register',height=2,width=20,command=register).pack()
    Label(text='').pack()
    Button(text='Login',height=2,width=20,command=login).pack()
    
    window.mainloop()
main_func()
```
